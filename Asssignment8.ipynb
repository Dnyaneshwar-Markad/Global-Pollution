```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv('Global_Pollution_Analysis.csv')

# Check for missing values
print(df.isnull().sum())
df
```

```python
from sklearn.preprocessing import MinMaxScaler

# Normalize pollution indices (0-1 range)
scaler = MinMaxScaler()
pollution_cols = ['Air_Pollution_Index', 'Water_Pollution_Index', 'Soil_Pollution_Index']
df[pollution_cols] = scaler.fit_transform(df[pollution_cols])
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode 'Country' (optional: one-hot encoding for many categories)
le = LabelEncoder()
df['Country_Encoded'] = le.fit_transform(df['Country'])

# Keep 'Year' as numeric or bin into categories if needed
df.head()
```

```python
# Energy Efficiency Ratio
df['Energy_Efficiency'] = df['Energy_Recovered (in GWh)'] / df['Energy_Consumption_Per_Capita (in MWh)']

# Pollution Severity Categories
df['Air_Pollution_Level'] = pd.cut(df['Air_Pollution_Index'], 
                                   bins=[0, 0.3, 0.7, 1], 
                                   labels=['Low', 'Medium', 'High'])
df['Water_Pollution_Level'] = pd.cut(df['Water_Pollution_Index'], 
                                     bins=[0, 0.3, 0.7, 1], 
                                     labels=['Low', 'Medium', 'High'])
df
```

```python
# Phase 2: Apriori Algorithm for Association Rule Mining

# Convert features to transactional format (list of lists)
transactions = []
for _, row in df.iterrows():
    transaction = [
        f"Air_{row['Air_Pollution_Level']}",
        f"Water_{row['Water_Pollution_Level']}",
        f"Renewable_Energy_{'High' if row['Renewable_Energy (%)'] > 30 else 'Low'}",
        f"Industrial_Waste_{'High' if row['Industrial_Waste (in tons)'] > df['Industrial_Waste (in tons)'].median() else 'Low'}"
    ]
    transactions.append(transaction)
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Convert transactions to one-hot encoded DataFrame
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_apriori = pd.DataFrame(te_ary, columns=te.columns_)

# Find frequent itemsets
frequent_itemsets = apriori(df_apriori, min_support=0.1, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.2)
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].sort_values('lift', ascending=False))
```

```python
# Phase 3: Model Evaluation & Validation

import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph from rules
G = nx.DiGraph()
for _, rule in rules.iterrows():
    G.add_edge(str(rule['antecedents']), str(rule['consequents']), weight=rule['lift'])

# Plot the graph
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=2000, font_size=10, 
        edge_color='gray', width=[d['weight']*0.1 for (u, v, d) in G.edges(data=True)])
plt.title("Association Rules (Edge Width = Lift)", fontsize=14)
plt.show()
```
