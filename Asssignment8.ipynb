```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv('Global_Pollution_Analysis.csv')

# Check for missing values
print(df.isnull().sum())
df
```

```python
from sklearn.preprocessing import MinMaxScaler

# Normalize pollution indices (0-1 range)
scaler = MinMaxScaler()
pollution_cols = ['Air_Pollution_Index', 'Water_Pollution_Index', 'Soil_Pollution_Index']
df[pollution_cols] = scaler.fit_transform(df[pollution_cols])
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode 'Country' (optional: one-hot encoding for many categories)
le = LabelEncoder()
df['Country_Encoded'] = le.fit_transform(df['Country'])

# Keep 'Year' as numeric or bin into categories if needed
df.head()
```

```python
# Energy Efficiency Ratio
df['Energy_Efficiency'] = df['Energy_Recovered (in GWh)'] / df['Energy_Consumption_Per_Capita (in MWh)']

# Pollution Severity Categories
df['Air_Pollution_Level'] = pd.cut(df['Air_Pollution_Index'], 
                                   bins=[0, 0.3, 0.7, 1], 
                                   labels=['Low', 'Medium', 'High'])
df['Water_Pollution_Level'] = pd.cut(df['Water_Pollution_Index'], 
                                     bins=[0, 0.3, 0.7, 1], 
                                     labels=['Low', 'Medium', 'High'])
df
```

```python
# Phase 2: Apriori Algorithm for Association Rule Mining
from mlxtend.frequent_patterns import apriori, association_rules

# Bin Energy_Consumption_Per_Capita and Energy_Recovered into 3 categories each
df['Energy_Consumption_Level'] = pd.qcut(df['Energy_Consumption_Per_Capita (in MWh)'], 
                                         q=3, 
                                         labels=['Low', 'Medium', 'High'])
df['Energy_Recovery_Level'] = pd.qcut(df['Energy_Recovered (in GWh)'], 
                                      q=3, 
                                      labels=['Low', 'Medium', 'High'])

# Also bin Soil_Pollution_Index if you want (already normalized 0-1)
df['Soil_Pollution_Level'] = pd.cut(df['Soil_Pollution_Index'], 
                                    bins=[-0.01, 0.33, 0.66, 1.0], 
                                    labels=['Low', 'Medium', 'High'])
df
```

```python
# Build the transactional format: One-hot encode the categorical features for Apriori
cols_for_apriori = [
    'Air_Pollution_Level', 'Water_Pollution_Level', 'Soil_Pollution_Level',
    'Energy_Consumption_Level', 'Energy_Recovery_Level'
]
df_apriori = pd.get_dummies(df[cols_for_apriori])

# Run Apriori algorithm
frequent_itemsets = apriori(df_apriori, min_support=0.1, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1.0)

# Show the top rules, sorted by confidence
rules.sort_values(by='confidence', ascending=False, inplace=True)
rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(10)
```

```python
# Filter only high pollution levels
high_pollution_df = df[
    (df['Air_Pollution_Level'] == 'High') |
    (df['Water_Pollution_Level'] == 'High') |
    (df['Soil_Pollution_Level'] == 'High')
]

apriori_data = pd.get_dummies(high_pollution_df[cols_for_apriori])

# --- Run Apriori algorithm ---
min_support = 0.1      # At least 10% of the filtered dataset
min_confidence = 0.6   # At least 60% confidence

frequent_itemsets = apriori(apriori_data, min_support=min_support, use_colnames=True)
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=min_confidence)

# --- Focus on rules where antecedent is a High Pollution Level and consequent is an Energy Recovery Level ---
def is_high_pollution(antecedents):
    return any(
        str(item).startswith('Air_Pollution_Level_High') or
        str(item).startswith('Water_Pollution_Level_High') or
        str(item).startswith('Soil_Pollution_Level_High')
        for item in antecedents
    )

def is_energy_recovery(consequents):
    return any(str(item).startswith('Energy_Recovery_Level') for item in consequents)

filtered_rules = rules[
    rules['antecedents'].apply(is_high_pollution) &
    rules['consequents'].apply(is_energy_recovery)
]

# Display the association rules
filtered_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# --- Visualize Frequent Itemsets (Top 10) ---
top_itemsets = frequent_itemsets.nlargest(10, 'support')
plt.figure(figsize=(10, 6))
plt.barh([' & '.join(itemset) for itemset in top_itemsets['itemsets']], top_itemsets['support'], color='skyblue')
plt.xlabel('Support')
plt.title('Top 10 Frequent Itemsets')
plt.gca().invert_yaxis()
plt.show()

# --- Visualize Association Rules as a Graph (Top 10 by confidence) ---

# Select top N rules
top_rules = filtered_rules.nlargest(10, 'confidence')

# Build the graph
G = nx.DiGraph()
for _, row in top_rules.iterrows():
    for ant in row['antecedents']:
        for cons in row['consequents']:
            G.add_edge(str(ant), str(cons), weight=row['confidence'], support=row['support'])

# Draw graph
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, k=0.8, seed=42)
edge_labels = {(u, v): f"{d['weight']:.2f}" for u, v, d in G.edges(data=True)}
nx.draw(G, pos, with_labels=True, node_color='lightgreen', node_size=2000, font_size=10, arrowsize=20)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')
plt.title('Association Rules (Confidence as edge label)')
plt.show()
```

```python
# # # Phase 4 - Reporting and Insights
# # Comparison: CNN Models for Delivery Prediction vs. Apriori for Association Rule Mining

# ## 1. Performance Comparison

# ### CNN Models for Delivery Prediction

# - **Strengths:**
#   - Excels at extracting complex, nonlinear patterns from image data (e.g., package images, road conditions, delivery environments).
#   - Can leverage spatial and pixel-level information that is inaccessible to traditional feature engineering methods.
#   - Suitable for real-time or automated prediction tasks where image cues are critical.

# - **Weaknesses:**
#   - Requires large, labeled image datasets and significant computational resources (GPUs/TPUs).
#   - Results are often “black box”—interpretability is low unless visualization techniques (e.g., Grad-CAM) are used.
#   - Prone to overfitting if data is insufficient or not diverse.

# - **Performance Metrics:**
#   - Evaluated using accuracy, precision/recall, F1-score, RMSE, etc. depending on the specific prediction target (e.g., on-time delivery, estimated delivery time).

# ### Apriori for Association Rule Mining

# - **Strengths:**
#   - Uncovers statistically significant and interpretable patterns in categorical/tabular data.
#   - Produces human-readable rules (e.g., "If Air Pollution is High, then Energy Recovery is Low").
#   - Useful for descriptive analytics and hypothesis generation.

# - **Weaknesses:**
#   - Cannot handle image data or extract features from raw pixels.
#   - Only finds associations, not causality or predictions.
#   - Suffers from combinatorial explosion with high-dimensional data or low support thresholds; may miss rare but important patterns if support is set too high.

# - **Performance Metrics:**
#   - Rule quality is measured by support, confidence, and lift.
#   - Statistical significance is ensured by minimum thresholds for these metrics.

# ---

# ## 2. Effectiveness of Feature Types

# ### Image-Based Features for CNN

# - **Effectiveness:**
#   - CNNs automatically learn visual features relevant to delivery prediction, such as package damage, road blockage, or weather.
#   - Useful when image cues directly impact delivery outcomes (e.g., predicting delays from traffic camera images).
#   - The quality and diversity of the image data directly affect the model’s performance.

# - **Limitations:**
#   - Not interpretable without additional tools.
#   - Requires expert tuning and large dataset.

# ### Rule-Based Learning in Apriori

# - **Effectiveness:**
#   - Best suited for structured, categorical/tabular data (e.g., pollution levels, energy recovery types).
#   - Rules are transparent and actionable, easy for domain experts to interpret and use in policy or business decisions.
#   - Does not require labeled outcomes, suitable for exploratory data analysis.

# - **Limitations:**
#   - Cannot capture complex, nonlinear, or pixel-level patterns.
#   - Requires discretization of continuous features, which may lead to information loss.

# ---

# ## 3. Conclusion

# - **CNNs** are powerful for prediction tasks where image content is critical, but they are resource-intensive and less interpretable.
# - **Apriori** is excellent for mining transparent, actionable rules in structured data, but not for prediction or image-based tasks.
# - **Complementarity:** In some advanced pipelines, CNN features can be extracted and then used as input for rule-based mining, combining the strengths of both approaches.
```
